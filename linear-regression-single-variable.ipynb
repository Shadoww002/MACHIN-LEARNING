```python
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/kaggle/input/per-capita-income/canada_per_capita_income.csv")
```

```python
df.head()
```

```python
# %matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df.per_capita_income,color='red',marker='+')
plt.show()
                                                                         
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[["year"]],df.per_capita_income)
```

```python
year_data= pd.read_csv("/kaggle/input/year-dataset/year.csv")
year_data.head()
```

```python
p = reg.predict(year_data)
```

```python
year_data["price"]= p
```

```python
year_data.head()
```

```python
year_data.to_csv("year_data_prediction.csv",index=False)

```
