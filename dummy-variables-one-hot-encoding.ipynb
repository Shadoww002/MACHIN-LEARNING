```python
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/kaggle/input/home-prices/homeprices.csv")
df.head()
```

```python
dummies = pd.get_dummies(df.town).astype(int)
dummies 
```

```python
merged = pd.concat([df,dummies] , axis = "columns")
merged
```

```python
final = merged.drop(["town","west windsor"],axis="columns")
final
```

```python
reg = linear_model.LinearRegression()

x = final.drop(["price"],axis="columns")
y = final["price"]

model = reg.fit(x,y)
```

```python
guess = pd.DataFrame([[2800,0,1],[4000,1,0]],columns=["area","monroe township","robinsville"])
result = model.predict(guess)
print(f"The Price of house with area = 2800 , in monroe township is :",result[0])
```
