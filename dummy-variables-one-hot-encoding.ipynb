```python
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/kaggle/input/home-prices/homeprices.csv")
df.head()
```

```python
dummies = pd.get_dummies(df.town).astype(int)
dummies 
```

```python
merged = pd.concat([df,dummies] , axis = "columns")
merged
```

```python
final = merged.drop(["town","west windsor"],axis="columns")
final
```

```python
reg = linear_model.LinearRegression()

x = final.drop(["price"],axis="columns")
y = final["price"]

model = reg.fit(x,y)
```

```python
guess = pd.DataFrame([[2800,0,1],[4000,1,0]],columns=["area","monroe township","robinsville"])
result = model.predict(guess)
print(f"The Price of house with area = 2800 , in monroe township is :",result[0])
```

```python
model.score(x,y)
```

```python
d = pd.read_csv("/kaggle/input/car-prices/carprices.csv")
d
```

```python
dummy = pd.get_dummies(d["Car Model"]).astype(int)
dummy
```

```python
combine = pd.concat([d,dummy],axis="columns")
combine
```

```python
final_df = combine.drop(["Car Model","Audi A5"],axis="columns")
final_df
```

```python
reg2 = linear_model.LinearRegression()

U = final_df.drop(["Sell Price($)"],axis="columns")
V = final_df["Sell Price($)"]

model2 = reg2.fit(U,V)
```

```python
Record = pd.DataFrame([[45000,4,0,1],[86000,8,0,1]],columns=["Mileage","Age(yrs)","BMW X5","Mercedez Benz C class"])
Prediction = model2.predict(Record)
print(f"Price of a mercedez benz that is 4 yr old with mileage 45000  :",Prediction[0])
print(f"Price of a BMW X5 that is 7 yr old with mileage 86000  :",Prediction[1])
```

```python
model2.score(U,V)
```
