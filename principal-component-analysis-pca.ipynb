```python
from scipy import stats
```

```python
df = pd.read_csv("/kaggle/input/heart-failure-prediction/heart.csv")
df.head()
```

```python
print(df.describe)
```

```python
df['Z_score'] = np.abs(stats.zscore(df[['Cholesterol']]))
```

```python
# Set threshold (commonly 3)
threshold = 3

# Remove outliers
df_clean = df[df['Z_score'] < threshold]
df_clean.head()
```

```python
df_clean.describe()
```

```python
df['Z_score'] = np.abs(stats.zscore(df[['RestingBP']]))
df_clean = df[df['Z_score'] < threshold]
```
